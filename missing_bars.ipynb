```python
import matplotlib.pyplot as plt

# 10개짜리 bar graph
plt.bar(range(1, 11), range(1, 11))
plt.show()
```

```python
# 100개짜리 bar graph
plt.bar(range(1, 101), range(1, 101))
plt.show()
```

```python
# 100개짜리 bar graph - Figure 폭 바꿈
plt.figure(figsize=(5, 5))
plt.bar(range(1, 101), range(1, 101))
plt.show()
```

```python
# bar width 조절
plt.bar(range(1, 101), range(1, 101), width=1)
plt.show()
```

```python
# bar width를 매우 두껍게 하면?
plt.bar(range(1, 101), range(1, 101), width=10)
plt.show()
```

```python
# bar width를 매우 가늘게 하면?
plt.bar(range(1, 101), range(1, 101), width=.1)
plt.show()
```

```python
x = range(100)
y = range(100)

fig, ax = plt.subplots(1, 4, figsize=(19, 5))
widths = [.1, .2, .3, .4]
for i, w in enumerate(widths):
    ax[i].bar(x, y, width=w)
plt.show()
```

```python
# 데이터 가져옴
!wget http://bit.ly/fruits_300_data -O fruits_300.npy
import numpy as np
fruits = np.load('fruits_300.npy')

# 사과 이미지 평균
apple = fruits[:100]
apple_mean = np.mean(apple, axis=0)

import matplotlib.pyplot as plt
plt.imshow(apple_mean, cmap='gray')
plt.show()
```

```python
# (10000,)으로 만들고 bar graph
# 책(혼공머신)과 정확히 같은 그림을 만들기 위해 subplots(1, 3)을 그대로 이용
apple_1d = apple_mean.reshape(-1, 100*100)[0]
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].bar(range(10000), apple_1d)
plt.show()
```

```python
# 동일 데이터의 scatter plot
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].scatter(range(10000), apple_1d, s=1)
plt.show()
```

```python
# axes 폭 바꾸고 다시 그려보기
plt.figure(figsize=(20, 5))
plt.bar(range(10000), apple_1d)
plt.show()
```

```python
# width=1로 바꾸고 다시 그려보기
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
axs[0].bar(range(10000), apple_1d, width=1)
plt.show()
```
